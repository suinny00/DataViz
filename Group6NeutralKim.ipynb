```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('store_data.csv', encoding='ISO-8859-1')
df
```

```python
costa_mesa_data = df[df['City'] == 'Costa Mesa']
costa_mesa_data
```

```python
x = costa_mesa_data['Category']
y = costa_mesa_data['Sales']

plt.xlabel('Category', fontsize = 16)
plt.ylabel('Sales', fontsize = 16)
plt.bar(x, y, color='yellow')

plt.show
```

```python
costa_mesa_data['Order Date'] = pd.to_datetime(costa_mesa_data['Order Date'])
costa_mesa_data = costa_mesa_data.sort_values('Order Date')

sales_by_OrderDate = costa_mesa_data.groupby('Order Date')['Sales'].sum()

plt.plot(sales_by_OrderDate.index, sales_by_OrderDate.values, marker='o')

plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Total Sales in Costa Mesa by Order Date')

plt.show()
```

```python
df['avg_discount_per_quantity'] = df['Discount'] / df['Quantity']

df['avg_discount_per_quantity'] = df['Discount'] / df['Quantity']

discount_data = df.groupby('avg_discount_per_quantity')['Discount'].mean().reset_index()

plt.bar(discount_data['avg_discount_per_quantity'], discount_data['Discount'])

plt.xlabel('Average Discount per Quantity Amount')
plt.ylabel('Average Discount')
plt.title('Average Discount by Quantity Amount in Costa Mesa')

plt.show()
```
