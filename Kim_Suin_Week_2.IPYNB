```python
import matplotlib.pyplot as plt
x1, y1 = [2,5], [4,4]
x2, y2 = [2,5], [7,7]
x3, y3 = [2,2], [4,7]
x4, y4 = [5,5], [4,7]
plt.plot(x1, y1, x2, y2, x3, y3, x4, y4, marker = '+', color="black")
plt.show()
```

```python
import matplotlib.pyplot as plt
   
country = ['A', 'B', 'C', 'D', 'E']
total_land_area = [5000, 43000, 74000, 25000, 39000]

plt.bar(country, total_land_area)
plt.title('Country Vs Total Land Area')
plt.xlabel('Country')
plt.ylabel('Total Land Area')
plt.show()
```
